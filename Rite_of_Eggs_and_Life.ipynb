Making a Modular Grid

```python
class MonsterDungeon():
    def __init__(self, rows, cols, lives):
        self.rows = rows
        self.cols = cols
        self.lives = lives
        
    def showGrid(self, player, monster, door, egg,): 
        #make sure you have acccess to coordinates in player and monster classes
        print('Player Coords: {}'.format(player.coords))
        print('Monster Coords: {}'.format(monster.coords))
        print('Door Coords: {}'.format(door.coords))
        print('Egg Coords: {}'.format(egg.coords))
        #print('Eggb Coords: {}'.format(eggb.coords))
        #TODO: ask about door coords issue "IndexError: list index out of range"
        print('Lives Left: {}'.format(self.lives))
        
        for i in range(self.rows):
            #print top border for each row
            print(' ---'*self.cols)
           
            for j in range(self.cols):
                
                if i == player.coords[1] and j == player.coords[0] and j == self.cols - 1:
                    print('| p ', end='|')
                elif i == player.coords[1] and j == player.coords[0]:
                    print('| p ', end='')
                elif i == door.coords[1] and j == door.coords[0] and j == self.cols - 1:
                    print('| d ', end='|')
                elif i == door.coords[1] and j == door.coords[0]:
                    print('| d ', end='')
                elif i == monster.coords[1] and j == monster.coords[0] and j == self.cols - 1:
                    print('| m ', end='|')
                elif i == monster.coords[1] and j == monster.coords[0]:
                    print('| m ', end='')
                elif i == egg.coords[1] and j == egg.coords[0] and j == self.cols - 1:
                    print('| e ', end='|')
                elif i == egg.coords[1] and j == egg.coords[0]:
                    print('| e ', end='')
                #elif i == eggb.coords[1] and j == eggb.coords[0] and j == self.cols - 1:
                    #print('| e ', end='|')
                #elif i == eggb.coords[1] and j == eggb.coords[0]:
                    #print('| e ', end='')
                elif j == self.cols-1:
                    print('|   ', end='|')
                else:
                    print('|   ', end= '')
                
                #start new row
            print('')
            
            if i == self.rows -1: 
                #print bottom border for each row
                print(' ---'*self.cols)
            
    def checkCollision (self, player, monster):
        if player.coords == monster.coords:
            self.lives -= 1
            #TODO: print statement below gets cleared, move elsewhere
            #print('You just got eaten! you have {} lives left.'.format(self.lives))
            
            
    def checkLoseCondition(self):
        if self.lives <= 0:
            return True
        
    def checkWinCondition(self, player, door):
        if player.coords == door.coords:
            return True

# define our global variables to be used
rows = 5
cols = 5
lives = 3
game = MonsterDungeon(rows, cols, lives)

#show the grid
```

Class Creation for Monsters, Player, Egg & Door

```python
class Player():
    def __init__(self, name):
        self.name = name
        self.coords = [0, 0] # x, y
    
    def movePlayer(self):
        while True:
            print('Type quit if you would like to stop playing!')
            ans = input('Move left/right/up/down?')

            #move player based on ans taken in
            if ans.lower() == 'quit':
                print('Thanks for playing!')
                return True
            elif ans.lower() == 'up':
                self.coords[1] -= 1
                break
            elif ans.lower() == 'down':
                self.coords[1] += 1
                break
            elif ans.lower() == 'left':
                self.coords[0] -= 1
                break
            elif ans.lower() == 'right':
                self.coords[0] += 1
                break
            else:
                print('Incorrect input, try again...')
                
                #clear output after asking question
                clear_output()


                
                
                
                
                
class Monster():
    def __init__(self):
        self.coords = [2,2] # x, y
        
    def moveMonster(self, cols, rows):
        self.coords = [random.randint(0, cols -1), random.randint(0, rows -1)]

class Door():
    def __init__(self):
        self.coords = [0,1]
        
    def moveDoor (self):
        curX = self.coords[0]
        curY = self.coords[1]
        self.coords = [random.randint(curX - 1, curX + 1), random.randint(curY - 1, curY + 1)]
        
class Egg():
    def __init__(self):
        self.coords = [3, 4] # x, y
        
    def moveEgg (self):
        self.coords = [random.randint(0, 1), random.randint(0, 1)]
        
    
    
    
```

Main Loop

```python
from IPython.display import clear_output
import random


#START OF MAIN LOOP
while True:
    
    # define our global variables to be used
    rows = 5
    cols = 5
    lives = 3
    game_over = False
    player = Player('Darius')
    monster = Monster()
    door = Door()
    egg = Egg()
    #eggb = Egg()
    game = MonsterDungeon(rows, cols, lives)

    #game loop
    while game_over != True:
        #clear output
        clear_output()

        #show the grid
        game.showGrid(player, monster, door, egg,)
        
        
        #call player to move
        game_over = player.movePlayer()
        
        #call monster to move, send in cols rows
        monster.moveMonster(cols, rows)
        
        #call door to move
        door.moveDoor()
        
        #call egg to move
        egg.moveEgg()
        
        # check to see if monster ate player
        game.checkCollision(player, monster)

        #check if lives are gone, then break
        #breaking because we don't want any other lines below this running
        if game.checkLoseCondition():
            #clear_output()
            print('You lost all your lives, better luck next time!')
            game_over = true
        elif game.checkWinCondition(player, door):
            print('Congratulations, you beat the monster!')
            game_over = True


    

    #ask if they want to play again, if not break out of while loop
    ans = input('Would you like to play again(yes/no)?')
    
    if ans.lower() == 'no':
        print('Thanks for playing!')
        break
```
